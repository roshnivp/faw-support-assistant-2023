```python
def filter_content(element):
        if '_expandable' in element:     
             del element['_expandable']
        if 'representation' in element:     
             del element['representation']  
        return element;
```

```python
data =  df.copy()
```

```python
from bs4 import BeautifulSoup

def extract_html(html):
    soup = BeautifulSoup(html,"html.parser")
    data_dict = {}
    for item in soup.select("h2"):
        header = item.get_text(strip=True)
        content = []
        for i in item.next_siblings:
            if i.name=="h2": break
            content.extend([x for x in i.stripped_strings])
            data_dict[header] = content
    return data_dict
    

```

```python
for i in data.index:
    newVal = extract_html(data['body'][i])
    data.at[i, 'body'] = newVal
print(data['body'][10])  
```

```python
res=[]
for i in data.index:
    content = data['body'][i]
    problem = 'NA'
    error = 'NA'
    solution = 'NA'
    if 'Problem' in content: 
        problem = content['Problem']
    if 'Symptom/Error Message' in content: 
        error = content['Symptom/Error Message']
    if 'Solution' in content: 
        solution = content['Solution']
    res.append(
        {
            'problem': problem,
            'error': error,
            'solution':solution,
            'title':data['title'][i]
        }
    )
print(res)   
```

```python
df_from_res = pd.DataFrame(res)
```

```python
df_from_res.head()
```

```python
merged_df=  pd.merge(df_from_res, df, on="title")
```
