```python
def filter_commments(comments):
    filtered_list = [
        dictionary for dictionary in comments
            if dictionary['author'].get('emailAddress') != 'no-reply@oracle.com'
    ]
    for element in filtered_list:
        if 'self' in element:     
             del element['self']     
        if 'id' in element:     
             del element['id']     
        if 'author' in element:     
            del element['author'] 
        if 'updateAuthor' in element:     
            del element['updateAuthor'] 
        if 'created' in element:     
            del element['created']
        if 'updated' in element:     
            del element['updated']
    return filtered_list;
```
